```python
# to display image in different window using cv2
while True:
    cv2.imshow('result',img)
    if cv2.waitKey(1)==27:
        break
cv2.destroyAllWindows()
```

```python
# to capture video input from camera and display it in different window using cv2
capture = cv2.VideoCapture(0)
while True:
    flag, img = capture.read()
    if flag:
        cv2.imshow('result',img)
    if cv2.waitKey(1)==27:
        break
capture.release()
cv2.destroyAllWindows()
```

```python
#  CascadeClassifier to Classify haarcascades features
haar_data = cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
```

```python
# detecting haarcascades features
haar_data.detectMultiScale(img)
```

```python
# to draw rectangle around haarcascades features in a image
while True:
    faces= haar_data.detectMultiScale(img)
    for x,y,w,h in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,255),4)
    cv2.imshow('result',img)
    if cv2.waitKey(1)==27:
        break
cv2.destroyAllWindows()
```

```python
# to draw rectangle around haarcascades features in a real time video
capture = cv2.VideoCapture(0)
while True:
    flag, img = capture.read()
    if flag:
        faces = haar_data.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,255),4)
            face = img[y:y+h , x:x+w, :]
            face = cv2.resize(face,(50,50))
            
        cv2.imshow('result',img)
        if cv2.waitKey(1)==27:
            break
capture.release()
cv2.destroyAllWindows()
```
